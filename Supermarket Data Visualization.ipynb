```python
#tarihleri incelemek için
#tarihler str halindedir bunu tarihe çeviricez
# - yi görünce str diye anlıyor
df['Date']=pd.to_datetime(df['Date'])
```

```python
df['day']=(df['Date']).dt.day
df['month']=(df['Date']).dt.month
df['year']=(df['Date']).dt.year
```

```python
df['month_name']=df['Date'].dt.month_name()
```

```python
df['weekday']=df['Date'].dt.day_name()
```

```python
df['Time']=pd.to_datetime(df['Time'])
```

```python
df['Hour']=(df['Time']).dt.hour
```

```python
sns.countplot(df['month_name'])
```

```python
plt.figure(figsize=(8,6))
#8 en 6 boy
plt.title('Monthly transaction')
sns.countplot(df['month_name'])
#countplot kasadan geçen müşterileri sayar
plt.xticks(rotation=30)
#90 dersen yazılar dik durur bu yazıların eğimini belirler
#aylık satış rakamlarını görmek istiyoruz
```

```python
#müşterlerin kaç tanesi kadın kaç tanesi erkek
plt.figure(figsize=(8,6))
plt.title('Male vs Female Customers')
sns.countplot(df['Gender']);
#countplot saydıklarını grafik yap demek
```

```python
#hangi ay yüzde toplam satışın kaçını yapmışız
plt.figure(figsize=(8,8))
explode=(0.1,0.05,0.05)
#explode parçaları ayırıyor.yani en büyük parçayı daha fazla ayırması için
df['month_name'].value_counts().plot.pie(autopct='%1.1f%%',startangle=60,explode=explode)
#pie sayesinde grafik bu şekilde
plt.title('Transaction per month')
```

```python
plt.figure(figsize=(8,8))
explode=(0.1,0.05,0.05,0.1,0.1,0.1)
#explode en büyük pastanın diğerlerinden ayrılmasını sağlıyor
df['Product line'].value_counts().plot.pie(autopct='%1.1f%%',startangle=60,explode=explode)
plt.title('Transaction per month')
#bütün satırların yüzdelerini gösteriyor
```
